```python
from numpy import random
x=random.binomial(n=10, p=0.5,size=10)
print(x)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.kdeplot(random.binomial(n=10,p=0.5,size=10))
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.displot(random.binomial(n=10,p=0.5,size=10),kde=False)
plt.show()
```

Iterating over array

```python
a=np.arange(0,45,5)
a=a.reshape(3,3)
a
```

```python
for x in np.nditer(a, order="C"):
    print(x)
```

```python
for x in np.nditer(a, order="F"):
    print(x)
```

Histogram

```python
import numpy as np
from matplotlib import pyplot as plt
a=np.array([1,2,3,4,6,7,10,20,40,50])
plt.hist(a, bins=[10,20,30,40,50])
plt.title("histogram")
plt.show()
```
